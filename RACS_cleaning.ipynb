```python
import pandas as pd
```

```python
input_file_path = "asu.tsv"  # Replace with the path to your input file
output_file_path = "RACS.csv"  # Replace with the path to your output file

with open(input_file_path, "r") as input_file:
    with open(output_file_path, "w") as output_file:
        lines_to_delete = 108
        lines_to_skip = [110,111]
        for i, line in enumerate(input_file, 1):
            if i > lines_to_delete:
                if i not in lines_to_skip:
                    output_file.write(line)

```
