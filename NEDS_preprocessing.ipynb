```python
import pandas as pd
```

```python
gs0 = pd.read_csv('Galaxy_Morphology_S0_Catalogue.csv')
gs0a = pd.read_csv('Galaxy_Morphology_S0a_Catalogue.csv') 
gs = pd.read_csv('Galaxy_Morphology_S_Catalogue.csv')
```

```python
gs = gs.iloc[:,1:]
```

```python
# list of data frames
dataframes = [gs, gs0, gs0a] 

# dictionary to save data frames
frames={} 

for key, value in enumerate(dataframes):    
  frames[key] = value # assigning data frame from list to key in dictionary
  print("key: ", key)
  print(frames[key], "\n")
```

```python
frames[0].head()
```

```python
for i in frames:
    frames[i] = frames[i][~frames[i]['Dec'].str.contains('\+')]
    print(frames[i].shape)
```

```python
for i in frames:
    # Convert the 'column_name' to a float
    frames[i]['Redshift'] = frames[i]['Redshift'].astype(float)

    # Filter rows where 'column_name' is less than or equal to 0.26
    frames[i] = frames[i][frames[i]['Redshift'] <= 0.26]
```

```python
for i in frames:
    print(frames[i].shape)
```

```python
NED_df = pd.concat(frames.values(), axis=0)
```

```python
NED_df.shape
```

```python
NED_df.reset_index(inplace = True,drop=True)
```

```python
NED_df.head()
```

```python
NED_df.to_csv('NED.csv')
```

```python
from datetime import datetime

# Function to convert RA string to time
def ra_str_to_time(ra_str):
    ra_str = ra_str.replace('h', ':').replace('m', ':').replace('s', '')
    time_obj = datetime.strptime(ra_str, '%H:%M:%S.%f')
    return time_obj.time()

# Apply the function to the RA column
NED_df['RA_time'] = NED_df['RA'].apply(ra_str_to_time)

# Display the updated DataFrame
print(NED_df[['Object Name', 'RA', 'RA_time']])
```

```python
# Function to convert RA from hms to degrees
def ra_to_deg(ra_str):
    # Split the string into hours, minutes, and seconds components
    h, m, s = map(float, ra_str.replace('h', '|').replace('m', '|').replace('s', '').split('|'))
    ra_deg = 15 * (h + m/60 + s/3600)   # Convert to degrees
    return ra_deg

# Apply the function to the 'RA' column and create a new 'RA_degrees' column
NED_df['RA_degrees'] = NED_df['RA'].apply(ra_to_deg)

# Display the DataFrame with the RA converted to degrees
NED_df.head()
```
